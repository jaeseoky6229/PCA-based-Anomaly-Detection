```python
from copulas.multivariate import GaussianMultivariate

# 각 결측률 데이터셋에 대해 모델을 따로 만듭니다.
model_01 = GaussianMultivariate()
model_02 = GaussianMultivariate()

# 각 환경의 분포를 추정(Fitting)합니다.
model_01.fit(residual1[feature_cols])
model_02.fit(residual2[feature_cols])

# 1. 각 열이 어떤 분포로 추정되었는지 확인 (0번 열 예시)
# 이 수치들을 비교하면 결측률에 따른 '잔차의 특성 변화'를 정량적으로 설명할 수 있습니다.
print("--- 0.1% Missing Rate Parameters (Column 0) ---")
print(model_01.univariates[0].to_dict())

print("\n--- 0.2% Missing Rate Parameters (Column 0) ---")
print(model_02.univariates[0].to_dict())

# 2. 전체적인 상관관계 구조(Correlation) 비교
print("\n--- Correlation Matrix comparison ---")
print(model_01.correlation - model_02.correlation) # 차이를 계산
```
